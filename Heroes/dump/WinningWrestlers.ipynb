```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in csv
df_2016 = pd.read_csv("WWE-Data-2016.csv")
#df_2016['year'] = '2016'
df_2016.rename(columns = {'Wins': 'Wins 2016', 'Losses': 'Losses 2016', 'Draws': 'Draws 2016' },inplace = True)
df_2016.head(5)
```

```python
# Load in csv
df_2015 = pd.read_csv("WWE-Data-2015.csv")
#df_2015['year'] = '2015'
df_2015.rename(columns = {'Wins': 'Wins 2015', 'Losses': 'Losses 2015', 'Draws': 'Draws 2015' },inplace = True)
df_2015.head(5)
```

```python
# Load in csv
df_2014 = pd.read_csv("WWE-Data-2014.csv")
#df_2014['year'] = '2014'
df_2014.rename(columns = {'Wins': 'Wins 2014', 'Losses': 'Losses 2014', 'Draws': 'Draws 2014' },inplace = True)
df_2014.head(5)
```

```python
# Load in csv
df_2013 = pd.read_csv("WWE-Data-2013.csv")
#df_2013['year'] = '2013'
df_2013.rename(columns = {'Wins': 'Wins 2013', 'Losses': 'Losses 2013', 'Draws': 'Draws 2013' },inplace = True)
df_2013.head(5)
```

```python
# Merge all of the CSV's
merged_df = pd.merge(df_2013, df_2014, on = "Wrestler", how="outer")
merged_df.head(5)
```

```python
merged_df.describe
```

```python
merged_df_2 = pd.merge(merged_df, df_2015, on = "Wrestler", how="outer")
merged_df_2.head(5)
```

```python
merged_df_2.describe
```

```python
merged_df_3 = pd.merge(merged_df_2, df_2016, on = "Wrestler", how="outer")
merged_df_3.head(5)
```
